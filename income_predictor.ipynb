```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
data['workclass'].head()
```

```python
data['workclass']=labelencoder.fit_transform(data['workclass'])
```

```python
data['workclass'].head()
```

```python
data['marital-status']=labelencoder.fit_transform(data['marital-status'])
data['education']=labelencoder.fit_transform(data['education'])
data['occupation']=labelencoder.fit_transform(data['occupation'])
data['income']=labelencoder.fit_transform(data['income'])
data['native-country']=labelencoder.fit_transform(data['native-country'])
```

```python
sns.heatmap(data.corr(),annot=True)
```

```python
x=data[['age','worklcass','occupation','marital-status','hours-per-week','native-country','education']]
y=data['income']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_x_train=scaler.fit_transform(x_train)
scaled_x_test=scaler.fit_transform(x_test)
```

```python
from sklearn.metrics import accuracy_score
def modelresult(prediction):
    print("Accuracy score of this model is:{}".format(accuracy_score(y_test,prediction)))
```

```python
from sklearn.linear_model import LogisticRegression
log_model=LogisticRegression()
```

```python
log_model.fit(scaled_x_train,y_train)
```

```python
y_pred=log_model.predict(scaled_x_test)
modelresult(y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
```

```python
knn=KNeighborsClassifier()
k_values=list(range(1,30))
pipe=Pipeline(steps=[('knn',knn)])
param_grid={"knn__n_neighbors":k_values}
cv_classifier=GridSearchCV(pipe,param_grid,cv=5,scoring="accuracy")
cv_classifier.fit(scaled_x_train,y_train.values.ravel())
```

```python
print(cv_classifier.best_params_)
```

```python
y_pred=cv_classifier.predict(scaled_x_test)
modelresult(y_pred)
```

```python
from sklearn.svm import SVC
svm=SVC()
params_grid_svc={"C":[0.1,0.5,1],"kernel":["linear","rbf","poly"]}
gridsvc=GridSearchCV(svm,params_grid_svc)
gridsvc.fit(scaled_x_train,y_train.values.ravel())
```

```python
y_pred=gridsvc.predict(scaled_x_test)
modelresult(y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfr_model=RandomForestClassifier()
n_estimator=[32,64,128]
max_features=[2,3,4,5]
bootstrap=[True,False]
oob_score=[True,False]
params_grid_rfr={"n_estimators":n_estimator,"max_features":max_features,"bootstrap":bootstrap,"oob_score":oob_score}
grid_rfr=GridSearchCV(rfr_model,params_grid_rfr)
grid_rfr.fit(scaled_x_train,y_train.values.ravel())
```

```python
print(grid_rfr.best_params_)
```

```python
y_pred=grid_rfr.predict(scaled_x_test)
modelresult(y_pred)
```
